# Using OneHotEncoder

```python
df=pd.read_csv('carprices.csv')
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first')
```

```python
dfle=df
```

```python
dfle
```

```python
dfle=dfle.rename(columns={
    "Car Model":"Model",
    "Mileage":"Mileage",
    "Sell Price($)":"Price",
    "Age(yrs)":"Age"
}
)
```

```python
encoded=ohe.fit_transform(dfle[['Model']])
```

```python
encoded=encoded.toarray()
```

```python
encoded
```

```python
encoded=pd.DataFrame(encoded)
```

```python
final_df=pd.concat([encoded,dfle.drop('Model', axis=1)],axis=1)
```

```python
final_df=final_df.rename(columns={0:"BMW X5",1:"Audi A5"})
```

```python
X=final_df.drop('Price',axis=1)
```

```python
y=final_df[['Price']]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,69000,6]])
```

```python
model.score(X,y)
```
