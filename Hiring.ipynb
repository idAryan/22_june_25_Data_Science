```python
import pandas as pd
```

```python
df=pd.read_csv('hiring.csv',header=None)
```

```python
df.columns=["experience","test_score","interview_score","salary"]
```

```python
df=df.drop(0,axis=0)
```

```python
df.head()
```

```python
row,col=df.shape
```

```python
df=df.fillna({
    'test_score':0
})
df['experience']=df['experience'].ffill().bfill()
```

# handling missing values

```python
df['test_score']=df['test_score'].astype(int)
```

```python
df
```

```python
df['test_score'].dtype
```

```python
import math
median_test_score=math.floor(df.test_score.median())
```

```python
median_test_score
```

```python
df['test_score']=df['test_score'].replace(0,median_test_score)
```

```python
word_map = {
    'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4,
    'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9,
    'ten': 10, 'eleven': 11, 'twelve': 12
}

df['experience']=df['experience'].map(word_map)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(df[['experience','test_score','interview_score']],df['salary'])
```

```python
reg.predict([[5,7,9]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
    pickle.dump(reg,f)
```
